```python
def read_in_data_txt(in_file):
    
    # in_file: point out the file name here/ in the main function

    data = pd.read_table(in_file, skiprows=4, nrows=50005,sep=':')
    
    pd.DataFrame(data,columns=list('condition value'))
    
    # Add column labels of the DataFrame
    data.columns=['number', 'condition', 'value']
    
    return data
```

```python
# Read-in data from ".txt" file

if __name__ == '__main__':

    in_file = 'large_dataset.txt'
    
    data_1 = read_in_data_txt(in_file)
    # call the function
    
    print(data_1)
    # see the read-in result
```

```python
def read_in_data_xls(in_file):
    
    # in_file: point out the file name here/ in the main function

    data = pd.read_excel(in_file, sheet_name = "Zugspitze-Monatsmittel")
    
    return data
```

```python
# Read-in data from ".xls" file

if __name__ == '__main__':

    in_file = 'Zugs_dataset.xls'
    
    data_2 = read_in_data_xls(in_file)
    # call the function
    
    print(data_2)
    # see the read-in result
```

```python
def read_in_data_csv(in_file):
    
    # in_file: point out the file name here/ in the main function

    data = pd.read_csv(in_file, skiprows=24, nrows=525)
    
    return data
```

```python
# Read-in data from ".csv" file

if __name__ == '__main__':

    in_file = 'Cape_dataset.csv'
    
    data_3 = read_in_data_csv(in_file)
    # call the function
    
    print(data_3)
    # see the read-in result
    
    print(data_3['DATE']) 
    # print only the column under [DATE], for checking the achievement of 'pandas.read_csv'
```
